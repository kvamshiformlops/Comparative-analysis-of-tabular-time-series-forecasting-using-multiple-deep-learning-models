```python
#Print the null values from all the columns
print(data.isnull().sum())
```

```python
#Data cleaning and pre-processing
data['pm2.5'].fillna(data['pm2.5'].median(), inplace=True)
print(data.isna().sum())
```

```python
data['datetime'] = pd.to_datetime(data[['year', 'month', 'day', 'hour']])
data.set_index('datetime', inplace=True)
data.drop(columns=['No', 'year', 'month', 'day', 'hour'], inplace=True)
```

```python
#Handling Categorical labels using Label encoder
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
categorical_columns = ['cbwd']
for column in categorical_columns:
    data[column] = label_encoder.fit_transform(data[column])
```

```python
#Outlier removal
from scipy.stats import zscore
z_scores = np.abs(zscore(data.select_dtypes(include=[np.number])))  # Only numeric columns
threshold = 3
outliers = (z_scores > threshold).any(axis=1)
data = data[~outliers]
print(data)
```

```python
#Normalize the data for better performance for the DL Models
from sklearn.preprocessing import MinMaxScaler
feature_columns = ['DEWP', 'TEMP', 'PRES', 'Iws', 'Is', 'Ir']  
target_column = 'pm2.5'

features = data[feature_columns]
target = data[[target_column]]

feature_scaler = MinMaxScaler()
target_scaler = MinMaxScaler()

features_scaled = feature_scaler.fit_transform(features)
target_scaled = target_scaler.fit_transform(target)

scaled_data = np.hstack((target_scaled, features_scaled))
```

```python
#Sequential modeling for time-series models
sequence_length = 24  # Number of time steps (e.g., 24 hours)
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i + seq_length, 1:])  # Use all features except the target as inputs
        y.append(data[i + seq_length, 0])    # Use the target value as the label
    return np.array(X), np.array(y)

X, y = create_sequences(scaled_data, sequence_length)

# Reshape X to add a feature dimension
X = X.reshape((X.shape[0], X.shape[1], X.shape[2]))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
print("\n The dimension of the training dataset are: ")
print(X.shape)
print("\n The dimensions of the test dataset are: ")
print(y.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
model = Sequential([
    LSTM(50, return_sequences=True,input_shape=(X_train.shape[1], X_train.shape[2])),
    Dropout(0.2),
    LSTM(50, return_sequences=False),
    Dropout(0.2),
    Dense(1)  # Single output for pm2.5 prediction
])
model.compile(optimizer=Adam(learning_rate=0.01), loss='mean_squared_error')
model.fit(X_train, y_train, epochs=20, batch_size=32, verbose=1)
```

```python
y_pred_lstm = model.predict(X_test)
```

```python
# Inverse transform the predictions and actual values
y_pred_original = target_scaler.inverse_transform(y_pred_lstm.reshape(-1, 1))
y_test_original = target_scaler.inverse_transform(y_test.reshape(-1, 1))
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, explained_variance_score

# Calculate evaluation metrics
mse = mean_squared_error(y_test_original, y_pred_original)
mae = mean_absolute_error(y_test_original, y_pred_original)
r2 = r2_score(y_test_original, y_pred_original)
evs = explained_variance_score(y_test_original, y_pred_original)
rmse = np.sqrt(mse)

# Print evaluation metrics
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"R² Score: {r2:.2f}")
print(f"Explained Variance Score (EVS): {evs:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
```

```python
from tensorflow.keras import layers, models

# Define the TCN model
class TCN(tf.keras.Model):
    def __init__(self, input_shape, filters=64, kernel_size=3, dilations=[1, 2, 4, 8]):
        super(TCN, self).__init__()
        self.tcn_layers = []
        for dilation in dilations:
            self.tcn_layers.append(layers.Conv1D(filters, kernel_size, padding='causal', dilation_rate=dilation, activation='relu'))
            self.tcn_layers.append(layers.Dropout(0.2))
        self.global_avg_pooling = layers.GlobalAveragePooling1D()
        self.output_layer = layers.Dense(1)

    def call(self, inputs):
        x = inputs
        for layer in self.tcn_layers:
            x = layer(x)
        x = self.global_avg_pooling(x)
        return self.output_layer(x)

# Instantiate and compile the model
input_shape = (24, 6)  # Example input shape: 24 time steps and 6 features
model_tcn = TCN(input_shape=input_shape)
model_tcn.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])

# Train the model
history = model_tcn.fit(X_train, y_train, epochs=20, batch_size=32)
```
